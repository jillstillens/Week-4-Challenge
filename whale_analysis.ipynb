# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_path = "Resources/whale_returns.csv"
whale_ret = pd.read_csv(whale_path,index_col="Date",parse_dates=True)
```

```python
# Count nulls
null_count = whale_ret.isnull().sum()
null_count
```

```python
# Drop nulls
whale_ret_valid = whale_ret.dropna()
null_count = whale_ret_valid.isnull().sum()
null_count
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns - note, there are two sets of data in this file, Algo 1 and Algo 2
algo_path = "Resources/algo_returns.csv"
algo_ret = pd.read_csv(algo_path,index_col="Date",parse_dates=True)
```

```python
# Count nulls
null_count = algo_ret.isnull().sum()
null_count
```

```python
# Drop nulls
algo_ret_valid = algo_ret.dropna()
null_count = algo_ret_valid.isnull().sum()
null_count
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices and sort into date order
sp_path = "Resources/sp500_history.csv"
sp_close = pd.read_csv(sp_path,index_col="Date",parse_dates=True)
sp_close = sp_close.sort_index()
```

```python
# Check Data Types
sp_close.dtypes
```

```python
# Fix Data Types to remove dollar sign and make field numeric (float)
sp_close[sp_close.columns[0:]] = sp_close[sp_close.columns[0:]].replace('[\$,]', '', regex=True).astype(float)
sp_close.head()
```

```python
# Calculate Daily Returns
sp_daily_returns = sp_close.pct_change()
sp_daily_returns
```

```python
# Drop nulls
sp_daily_ret_valid = sp_daily_returns.dropna()
null_count = sp_daily_ret_valid.isnull().sum()
null_count
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_daily_ret_valid = sp_daily_ret_valid.rename(columns={'Close':'S&P Returns'})
sp_daily_ret_valid.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined = pd.concat([whale_ret_valid, algo_ret_valid, sp_daily_ret_valid], axis="columns", join="inner")
# Rename columns to shorter names
combined.columns = ['Soros', 'Paulson', 'Tiger','Berkshire','Algo 1','Algo 2','SP500']
combined
```

### Determine which portfolios are riskier than the S&P 500

```python
# Calculate  the daily standard deviation of S&P 500, which will be considered baseline
sp_std = daily_std["SP500"]

#create a range which is considered baseline: add and subtract .000001 to create a baseline equivalent range
sp_std_low = sp_std - .000001
sp_std_high = sp_std + .000001

#create bins, to identify those which are less risky, more risky, or match baseline risk
bins = [0, sp_std_low,sp_std_high, 1]

# Create names for the bins
group_names = ["Less Risky", "Baseline Risk", "More Risky"]

# Determine which portfolios are riskier than the S&P 500
pd.cut(daily_std, bins, labels=group_names)
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = daily_std * np.sqrt(252)
annualized_std
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios - using SP500 as "risk-free rate" (ie: benchmark)
# Note that because I'm using SP500 instead of a true risk free rate, the sharpe ratios
# are useful in this context to compare each portfolio to the others; the sharpe ratio
# would not be useful to compare to other, external sharpe ratios.  In other words, the 
# sharpe ratio of -.34 for Soros below is not the same as a sharpe ratio of -.34 when compared
# to a true risk free rate.
sharpe_ratios = ((combined.mean() - combined['SP500'].mean()) * 252) / (combined.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot.bar()
```

### Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

Algo 1 looks attractive - the cumulative returns are higher than the SP500, the standard deviation is lower, indicating less risk; it's loosely correlated with the sp500 so provides more diversity, and the sharpe ratio is pretty good at .65 (relative to SP500, not risk free rate).
Algo 2, on the other hand, isn't quite as appealing - the cumulative returns are lower than the sp500, the standard deviation is approximately the same, and it's highly correlated with the SP500, but with a negative sharpe ratio.  All in all, it does not provide a good return for the risk.
The remaining portfolios don't seem particularly attractive - Tiger and Paulson in particular seem more risky than the returns would suggest.
